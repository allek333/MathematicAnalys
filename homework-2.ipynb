### Задание 1
Как соотносятся понятия “множество” и “последовательность”? 
(в ответе использовать слова типа: часть, целое, общее, частное, родитель, дочерний субъект и т. д.)

Под множеством следует понимать совокупность объектов или понятий объединённых в группу по какому-либо признаку. Последовательность - логически связанный набор элементов, изменяющийся по определённому закону.

Из данных определений понятно, что множество может быть частью последовательности и наоборот - последовательность может быть частью множества. Если закон изменения элементов принять за правило, по которому следует объединять элементы, эту последовательность можно назвать множеством. Если же в како-либо множестве упорядочить элементы по какому-либо закону, это множество можно будет назвать последовательностью.

#### Найти 12-й член заданной неявно последовательности

a(1) = 128, a(n+1) - a(n) = 6

```python
'''Для решения перепишем последовательность в привычном виде
a(n+1) - a(n) - 6 = b'''
a = 1
for n in range(129, 129+12):
    b = a*(n+1) - a*n - 6
    print(b)
```
